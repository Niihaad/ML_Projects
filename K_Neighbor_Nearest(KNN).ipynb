```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
from google.colab import files
uploaded = files.upload()
dataset= pd.read_csv("diabetes.csv")
```

```python
len(dataset)
dataset.head()
```

```python
# Replace zeroes
zero_not_accepted = ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI', 'Insulin']
```

```python
# replacing the 0 in the columns by the average
for column in zero_not_accepted:  # iterating through columns specified in zero_not_accepted
    dataset[column] = dataset[column].replace(0, np.NaN)  # replacing 0s with NaN in the specified column
    mean = int(dataset[column].mean(skipna=True))  # calculating the mean of the column, ignoring NaN values
    dataset[column] = dataset[column].replace(np.NaN, mean)  # replacing NaN values with the calculated mean
```

```python
#Data after removing zeros
len(dataset)
dataset.head()
```

```python
# split dataset
X = dataset.iloc[:, 0:8]
y = dataset.iloc[:, 8]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
# Feature scaling
#This practice ensures that all features contribute equally to the analysis, prevents dominance by features with larger scales,
# and helps certain algorithms converge faster and perform better by making them less sensitive to feature scales
# Creating a StandardScaler object
sc_X = StandardScaler()

# Scaling (standardizing) the features in the training set (X_train)
X_train = sc_X.fit_transform(X_train)
# Using fit_transform to standardize the training set features

# Scaling the features in the test set (X_test) using the same scaler computed from the training set
X_test = sc_X.transform(X_test)
# Using transform to apply the same scaling transformation to the test set as applied to the training set
```

```python
#after StandardScaller here all values will be between -1 and 1
print(X_test)
```

```python
# Define the model: Init K-NN
#p=2 because we have : with diabets or not (2 choices)
classifier = KNeighborsClassifier(n_neighbors=11, p=2,metric='euclidean')
```

```python
# Fit Model
classifier.fit(X_train, y_train)
```

```python
# Predict the test set results
y_pred = classifier.predict(X_test)
y_pred
```

```python
# Evaluate Model
cm = confusion_matrix(y_test, y_pred)
print (cm)
print(f1_score(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```
